```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#Fucntional Method
plt.plot(x,y)
plt.show() #If you are using some other IDE's
```

```python
plt.plot(x,y,'r-')
```

```python
plt.plot(x,y)
plt.xlabel('This is x axis')
plt.ylabel('This is y axis')
plt.title('This is Me')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x)
```

```python
# Object Oriented Method
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.0,0.1,0.5,0.8])
axes2 = fig.add_axes([0.6,0.1,0.5,0.8])

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title('I am 1')
axes2.set_title('I am 2')
```
