# Plot Appearence

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green',linewidth=2,alpha=0.5,linestyle='-',
        marker='o',markersize=20,markerfacecolor='yellow')  

# RGB HEx code can also be used
# linewidth = lw
# linestyles = --,-,steps,:
# marker = 'o','+' 
```
